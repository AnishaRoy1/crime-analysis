```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import folium
pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)
pd.set_option('display.width', 100)
pd.set_option('display.colheader_justify', 'left')
pd.set_option('display.precision', 5)
```

### Creating Header Dataframe

```python
df = pd.read_excel("2 Districtwise SLL Crimes_2021.xlsx")
df.iloc[0:4]
```

## Creating West Bengal Dataframe

```python
df1=df.loc[df['Districtwise SLL Crimes - 2020'] == 'State: West Bengal'].dropna()
df1['Unnamed: 2'] = df1['Unnamed: 2'].astype(int) 
s ='Unnamed: '
for i in range (2,92):
    y = str(i)
    e = s + y
    df1[e] = df1[e].astype(int)
df1
```

# Crimes related to women

```python
#Immoral Traffic
temp=df1.iloc[0:39]
df2 = temp.sort_values('Unnamed: 3', ascending=False)
df2 = df2.iloc[0:10]
sns.set(rc={'figure.figsize': (30, 7)})
ax = sns.barplot(x=df2['Unnamed: 1'],y=df2['Unnamed: 3'])
ax.set_xlabel('District', fontsize=20)
ax.set_ylabel('Number of Immoral Traffic Acts', fontsize=20)
list_murder = set(df2['Unnamed: 1'])
```

```python
temp=df1.iloc[0:39]
df2 = temp.sort_values('Unnamed: 9', ascending=False)
df2 = df2.iloc[0:10]
sns.set(rc={'figure.figsize': (30, 7)})
ax = sns.barplot(x=df2['Unnamed: 1'],y=df2['Unnamed: 9'])
ax.set_xlabel('District', fontsize=20)
ax.set_ylabel('Number of Prohibition of Child Marriage Acts', fontsize=20)
list_murder = set(df2['Unnamed: 1'])
```

```python
women_crime = df1.iloc[0:39]['Unnamed: 7'].sum()
children_crime = df1.iloc[0:39]['Unnamed: 13'].sum()
sc_crime = df1.iloc[0:39]['Unnamed: 18'].sum()
arms_crime = df1.iloc[0:39]['Unnamed: 30'].sum()
drug_crime = df1.iloc[0:39]['Unnamed: 46'].sum()
values = [women_crime, children_crime,sc_crime,arms_crime,drug_crime]
labels = ['total crime ACT related to women', 'total crime ACT related to children','total crime ACT related to SC-ST','total crime ACT related to arms', 'total crime ACT related to drugs' ]
plt.pie(values, labels=labels, autopct='%1.1f%%')
plt.title('Percentage of Women, Children, SC-ST, Arms, Drugs related crimes')
plt.show()
```

# Crimes related to children

```python
# Extract the values for each column
sexual_offences = df1.iloc[0:39]['Unnamed: 8'].sum()
pre_natal = df1.iloc[0:39]['Unnamed: 11'].sum()
juvenile_justice = df1.iloc[0:39]['Unnamed: 9'].sum()
child_marriage = df1.iloc[0:39]['Unnamed: 10'].sum()
values = [sexual_offences, pre_natal, juvenile_justice, child_marriage]
labels = ['crime ACT related to children sexual_offences','crime ACT related to children pre_natal' ,'crime ACT related to children juvenile_justice','crime ACT related children child_marriage']
plt.pie(values, labels=labels, autopct='%1.1f%%')
plt.title('Percentage of sexual_offences, pre_natal, juvenile_justice, child_marriage')
plt.show()
```

```python
cols = ['Unnamed: 31', 'Unnamed: 32', 'Unnamed: 33']
plt.figure(figsize=(6,3))
dataplot = sns.heatmap(df1[cols].corr(), annot=True)
ax = dataplot.axes
ax.set_xticklabels(['The Information Technology Act, 2000', 'The Copy Right Act, 1957', 'The Trade Marks Act, 1999'], rotation=45)
ax.set_yticklabels(['The Information Technology Act, 2000', 'The Copy Right Act, 1957', 'The Trade Marks Act, 1999'], rotation=0)
plt.show()
```
